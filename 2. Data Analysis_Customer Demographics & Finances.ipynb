```python
fig, axes = plt.subplots(4, 2, figsize=(10, 20))

stackedbar(axes[0, 0], df_yes, df_no, 'Age range', '\nAge Groups')
barcrdist(axes[0, 1], df_yes, df_no, 'Age range', '\nAge Groups')
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=70)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=70)
axes[0, 0].set_xlabel("Age Group")
axes[0, 1].set_xlabel("Age Group")

stackedbarorder(axes[1, 0], df_yes, df_no, 'Type of job', 'desc order', '\nJob Types')
barordercrdist(axes[1, 1], df_yes, df_no, 'Type of job', 'desc order', '\nJob Types')
axes[1, 0].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=70)
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=70)

stackedbarorder(axes[2, 0], df_yes, df_no, 'Marital status', 'desc order','\nMarital Status')
barorderdist(axes[2, 1], df_yes, 'Marital status', 'desc order', '\nMarital Status')

stackedbarorder(axes[3, 0], df_yes, df_no, 'Education', 'education', '\nEducation')
barordercrdist(axes[3, 1], df_yes, df_no, 'Education', 'education', '\nEducation')
axes[3, 0].set_xticklabels(axes[3, 0].get_xticklabels(), rotation=45)
axes[3, 1].set_xticklabels(axes[3, 1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
